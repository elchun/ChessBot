```python
# def createURDF(name, properties):
#     lines = []
#     lines.append(f'<robot name="{name}">\n')
#     lines.append(f'  <link name="piece_body">\n')


def make_parts(obj_filename, properties):
    section = 'General'  # General, BoundingBox, Inertia, NA
    MoI = {}
    CoM = None
    mass = None
    radius = 0
    sphere_radius = 0.001
    bb = {}
    for line in properties.split('\n'):
        if not line:
            continue
        if 'Bounding Box' in line:
            section = 'BoundingBox'
            continue
        if 'Moment of Inertia at Center of Mass' in line:
            section = 'Inertia'
            continue
        if 'Moment of Inertia at Origin' in line:
            section = 'NA'
            continue

        if section == 'General':
            values = line.split()
            if values[0] == 'Mass':
                mass = values[1]
                continue
            if 'Center of Mass' in line:
                CoM = [values[3], values[5], values[7]]
                continue
            continue

        if section == 'BoundingBox':
            values = line.split()
            bb[values[0]] = values[1]

        if section == 'Inertia':
            values = line.split()
            MoI[values[0]] = values[1]
    radius = max(float(bb['Length']), float(bb['Width'])) / 2

    # print(radius)
    # print(MoI)

    inertial = '<inertial>\n' \
    f'  <mass value="{mass}"/>\n' \
    f'  <origin xyz="{CoM[0]} {CoM[1]} {CoM[2]}" rpy="0 0 0"/>\n' \
    f'  <inertia ixx="{MoI["Ixx"]}" ixy="{MoI["Ixy"]}" ixz="{MoI["Ixz"]}" iyy="{MoI["Iyy"]}" iyz="{MoI["Iyz"]}" izz="{MoI["Izz"]}"/>\n' \
    '</inertial>\n'

    collision = '<collision>\n' \
    f'  <origin rpy="0 0 0" xyz="0 0 {float(bb["Height"]) / 2}"/>\n' \
    f'  <geometry>\n' \
    f'    <cylinder radius="{radius}" length="{bb["Height"]}"/>\n' \
    f'  </geometry>\n' \
    '</collision>\n'

    contact_rad = radius * 0.8
    contact = [collision]
    for i in range(3):
        x = np.cos(2 * np.pi/3 * i)*contact_rad
        y = np.sin(2 * np.pi/3 * i)*contact_rad
        x = round(x, 3)
        y = round(y, 3)

        sphere_collision = '<collision>\n' \
        f'  <origin rpy="0 0 0" xyz="{x} {y} 0"/>\n' \
        f'  <geometry>\n' \
        f'    <sphere radius="{sphere_radius}"/>\n' \
        f'  </geometry>\n' \
        '</collision>\n'
        contact.append(sphere_collision)
    
    collision = ''.join(contact)

    visual = '<visual>\n' \
    f'  <origin rpy="0 0 0"/>\n' \
    f'  <geometry>\n' \
    f'    <mesh filename="{obj_filename}"/>\n' \
    f'  </geometry>\n' \
    f'</visual>\n'

    return inertial, collision, visual

def make_board_parts(obj_filename, properties):
    section = 'General'  # General, BoundingBox, Inertia, NA
    MoI = {}
    CoM = None
    mass = None
    radius = 0
    bb = {}
    for line in properties.split('\n'):
        if not line:
            continue
        if 'Bounding Box' in line:
            section = 'BoundingBox'
            continue
        if 'Moment of Inertia at Center of Mass' in line:
            section = 'Inertia'
            continue
        if 'Moment of Inertia at Origin' in line:
            section = 'NA'
            continue

        if section == 'General':
            values = line.split()
            if values[0] == 'Mass':
                mass = values[1]
                continue
            if 'Center of Mass' in line:
                CoM = [values[3], values[5], values[7]]
                continue
            continue

        if section == 'BoundingBox':
            values = line.split()
            bb[values[0]] = values[1]

        if section == 'Inertia':
            values = line.split()
            MoI[values[0]] = values[1]
    radius = max(float(bb['Length']), float(bb['Width'])) / 2

    # print(radius)
    # print(MoI)

    inertial = '<inertial>\n' \
    f'  <mass value="{mass}"/>\n' \
    f'  <origin xyz="{CoM[0]} {CoM[1]} {CoM[2]}" rpy="0 0 0"/>\n' \
    f'  <inertia ixx="{MoI["Ixx"]}" ixy="{MoI["Ixy"]}" ixz="{MoI["Ixz"]}" iyy="{MoI["Iyy"]}" iyz="{MoI["Iyz"]}" izz="{MoI["Izz"]}"/>\n' \
    '</inertial>\n'

    collision = '<collision>\n' \
    f'  <origin rpy="0 0 0" xyz="0 0 {-float(bb["Height"]) / 2}"/>\n' \
    '  <geometry>\n' \
    f'    <box size="{bb["Length"]} {bb["Width"]} {bb["Height"]}"/>\n' \
    f'  </geometry>\n' \
    '</collision>\n'

    visual = '<visual>\n' \
    f'  <origin rpy="0 0 0"/>\n' \
    f'  <geometry>\n' \
    f'    <mesh filename="{obj_filename}"/>\n' \
    f'  </geometry>\n' \
    f'</visual>\n'

    return inertial, collision, visual

def make_URDF(name, inertial, collision, visual, piece=True):
    lines = []
    lines.append(f'<?xml version="1.0" encoding="utf-8"?>')
    lines.append(f'<robot name="{name}">')
    if piece:
        lines.append(f'  <link name="piece_body">')
    else:
        lines.append(f'  <link name="board_body">')
    for subpart in [inertial, collision, visual]:
        for line in subpart.split('\n'):
            if line:
                lines.append('    ' + line)
    lines.append(f'  </link>')
    lines.append('</robot>')
    return '\n'.join(lines)


```

```python
obj_filename = 'Bishop_B.obj'
obj_name = 'Bishop_B'
obj_prop = bishop_prop
urdf_fname = 'Bishop_B.urdf'

inertial, collision, visual = make_parts(obj_filename, obj_prop)
urdf = make_URDF(obj_name, inertial, collision, visual)

with open(urdf_fname, 'w') as f:
    f.write(urdf)
print(urdf)
```
