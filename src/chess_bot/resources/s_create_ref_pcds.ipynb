## Utility script to convert objs to reference pcds to use in icp

```python
import trimesh
import numpy as np

from plotly_utils import multiplot
```

```python
pieces = [
    'Bishop',
    'King',
    'Knight',
    'Pawn',
    'Queen',
    'Rook',
]

for piece in pieces:
    mesh = trimesh.load_mesh(f'../models/{piece}_B.obj', 'obj')
    pcd = np.array(mesh.sample(2000))
    # pcd = pcd[pcd > 0]
    np.save(f'reference_pcds/{piece}.npy', pcd)
```
